```python
suits=('Hearts','Diamonds','Spades','Clubs')
ranks=['2','3','4','5','6','7','8','9','10','Jack','Queen','King','Ace']
values=[2,3,4,5,6,7,8,9,10,10,10,10]


class BlackJack:

    def __init__(self,suits,ranks,values):
        self.suits=suits
        self.ranks=ranks
        self.values=values

    def CreatingDeck(self):
        deck=[]
        for suit in suits:
            for rank in ranks:
                deck.append(suit+' '+rank)
        return deck

    def SuffleDeck(self,deck):
        from random import shuffle
        shuffle(deck)
        return deck

    def Balance(self):
        while True:
            try:
                balance=int(input("Enter Amount you want to Buy : $"))
                if balance!=0:
                    return balance
            except:
                print("Try Again!!!")
        
    def Dealer(self,deck):
        print("Dealer : ")
        dealer=['None']
        dealer.append(deck.pop(0))
        dealer.append(deck.pop(0))
        print(f"{dealer[1]}")
        return [deck,dealer]

    def Player(self,deck):
        player=['None']
        player.append(deck.pop(0))
        player.append(deck.pop(0))
        print(f"{player[1]}")
        print(f"{player[2]}")
        return [deck,player]

    def Input(self):
        while True:
            decision=input("Type:\n\t0.Hit\n\t1.Stay:")
            if decision=='1':
                print("STAY!!!")
                break
            elif decision=='0':
                break
        return decision
        
    def Hit(self,deck,player):
        player.append(deck.pop(0))
        print("!!!HIT!!!")
        print(f"{player[3]}")
        return [deck,player]
    
    def Sum(self,player):
        sum=0
        for count,rank in enumerate(self.ranks[0:12]):
            for card in player:
                if rank in card:
                    sum+=self.values[count]
        for card in player:
            if 'Ace' in card:
                    sum+=1
        for card in player:
            if 'Ace' in card:
                if sum+10<22:
                    sum+=10
        return sum
    
    def TwentyOne(self,sum):
        if sum>21:
            return False

    def Print(self,dealer,player,balance,round,player_hit,dealer_hit,player_sum,dealer_sum):
        from IPython.display import clear_output
        clear_output()
        print(f"Balance : ${balance}")
        print(f"Round {round}\n")
        print("Dealer : ")
        print(f"{dealer[1]}")
        print(f"{dealer[2]}")
        if dealer_hit==1:
            print("!!!HIT!!!")
            print(f"{dealer[3]}")
        print(f"Dealer's hand : {dealer_sum}")
        print("\nPlayer : ")
        print(f"{player[1]}")
        print(f"{player[2]}")
        if player_hit==1:
            print("!!!HIT!!!")
            print(f"{player[3]}")
        print(f"Player's hand : {player_sum}")
    
    def DealerHit(self,player_sum,dealer_sum):
        if player_sum > dealer_sum:
            return True

    def Winner(self,player_sum,dealer_sum,balance,bet):
        if player_sum>dealer_sum:
            print("-----Player Wins-----")
            print(f"Gained ${bet}")
            balance+=bet
            print(f"Current Balance : ${balance}")
            return balance
        elif player_sum<dealer_sum:
            print("-----Dealer Wins-----")
            print(f"Lost ${bet}")
            balance-=bet
            print(f"Current Balance : ${balance}")
            return balance
        else:
            print("-----Draw-----")
            return balance

    def Proceed(self,round):
        while True:
            if input("Press 1 to Proceed")=='1':
                from IPython.display import clear_output
                clear_output()
                round+=1
                return round

    def PlaceBet(self,balance):
        while True:
            try: 
                bet=int(input("Place Bet : $"))
                if bet in range(1,balance+1):
                    print(f"Placed Bet : ${bet}\n")
                    return bet
            except:
                print("Try Again!!!")
```
