```python
class Calculator:
    
    def unpacking(self,string):
        import math
        num=0
        lst=[]
        for obj in string:
            if obj.isnumeric()==True:
                if lst==[]:
                    lst.append(int(obj))
                else:
                    try:
                        int(lst[num])
                        lst[num]=lst[num]*10+int(obj)
                        num-=1
                    except:
                        lst.append(int(obj))
            else:
                if obj=='e':
                    lst.append(math.e)
                elif obj=='i':
                    if lst[num-1]=='p':
                        lst[num-1]=math.pi
                        num-=1
                else:
                    lst.append(obj)
            num+=1
        return lst

    def find_large(self,lst,letter):
        for num,obj in enumerate(lst):
            if obj==letter:
                index=num
        if index==0:
            return len(lst)
        else:
            return index

    def find_small(self,lst,letter):
        return lst.index(letter)

    def operation(self,lst,index1,index2):
        for count in range(index1+1,index2):
            if lst[count]=='/':
                lst[count]=lst[count-1]/lst[count+1]

            elif lst[count]=='*':
                lst[count]=[count-1]*lst[count+1]
                
            elif lst[count]=='+':
                lst[count]=lst[count-1]+lst[count+1]

            elif lst[count]=='-':
                lst[count]=lst[count-1]-lst[count+1]
                
            lst.pop(count+1)
            lst.pop(count-1)
    
    def bracket(self,list,index1,index2):
        if lst[index1-1] not in ['+','-','*','/']:
            integer=False
            for obj in range(index1):
                try:
                    int(obj)
                    integer=True
                    break
                except:
                    pass
            if integer==True:
                lst[index1]='*'
            else:
                lst.pop(index1)
        if len(lst)-1==index2:
            lst.pop(index2)
        else:
            if lst[index2] not in ['+','-','*','/']:
                integer=False
                for obj in range(index2,len(lst)):
                    try:
                        int(obj)
                        integer=True
                        break
                    except:
                        pass
                if integer==True:
                    lst[index2]='*'
                else:
                    lst.pop(index2)
                
        return lst
```

```python
# Taking input (string)
string='(1+23(pi+6))'
# Calling Class
cal=Calculator()

# Unpacking (list)
lst=cal.unpacking(string)
print(lst)
print(len(lst))
# Finding '(' in lst
index1=cal.find_large(lst,'(')
index2=cal.find_small(lst,')')
print(index1)
print(index2)

# Checking Operator
lst=cal.operation(lst,index1,index2)
print(lst)

# Bracket Remove
index1=cal.find_large(lst,'(')
index2=cal.find_small(lst,')')
print(index1)
print(index2)
lst=cal.bracket(lst,index1,index2)
print(lst)
```
